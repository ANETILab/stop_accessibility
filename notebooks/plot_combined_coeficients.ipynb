```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("../output/regression/allModelFrame_a.csv")
```

```python
grid = sns.FacetGrid(
    data=data.drop(columns=["std_error", "model"]),
    hue="type",
    col="city",
    row="variable",
)
grid.map(plt.scatter, "coef", "type", marker="o")
```

```python
grid.axes[0][0].set_xlim([-0.05, 0.15])
grid.fig
```
